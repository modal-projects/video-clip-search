# Video Clip Search Demo

This notebook sends a text query to the deployed `/search` endpoint and renders the best matching video clip returned by the server.

```python
import requests
from IPython.display import Video, display


SEARCH_BASE_URL = "https://modal-labs-adamkelch-dev--video-clip-search-servers-vide-952e86.modal.run"
SEARCH_URL = f"{SEARCH_BASE_URL}/search"

def search_video(text_query: str) -> dict:
    payload = {
        "type": "text",
        "text": text_query,
    }
    response = requests.post(SEARCH_URL, json=payload, timeout=120)
    response.raise_for_status()
    return response.json()

```

```python
query = "A video of a dancer spinning like a ballerina"
result = search_video(query)
video_url = result["url"]
score = result["score"]

print(f"Query: {query}")
print(f"Best match score: {score:.4f}")
print(f"Video URL: {video_url}")

display(Video(url=video_url, embed=False, html_attributes="controls"))
```
