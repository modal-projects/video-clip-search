## Demo - Video Search App Running on Modal

This notebook sends a text query to the deployed `/search` Modal web endpoint and renders the best matching video clip returned by the server.

### Prerequisites

Before running this notebook, make sure the following steps have been completed:

**1. Run the embedding job**

The video embedding job must have been run (in detached mode so it continues in the background):

```bash
modal run --detach embed.py
```

This processes the video dataset and populates the vector index that the search server queries against.

**2. Deploy the search server**

The search server must be deployed so the notebook can reach its `/search` web endpoint:

```bash
modal deploy search_server.py
```

Once both steps are done, you can run the cells below to issue text queries and view the top-matching video clips.

````python
import modal
import requests
from IPython.display import Video, display

search_app_url = modal.Function.from_name("video-clip-search-server", "video_search_server").get_web_url()
SEARCH_URL = f"{search_app_url}/search"

def search_video(text_query: str) -> dict:
    payload = {
        "type": "text",
        "text": text_query,
    }
    response = requests.post(SEARCH_URL, json=payload, timeout=120)
    response.raise_for_status()
    return response.json()

````

````python
query = "A video of a dancer spinning like a ballerina"
result = search_video(query)
video_url = result["url"]
score = result["score"]

print(f"Query: {query}")
print(f"Best match score: {score:.4f}")
print(f"Video URL: {video_url}")

display(Video(url=video_url, embed=False, html_attributes="controls"))
````
